**Instructions**: To avoid getting the `ValueError`, we need to remove non-numeric attributes. To do this, we should use a ColumnSelector. We will use this so the model knows what columns to keep.

Relevant documentation:
* [`make_column_selector`](https://scikit-learn.org/stable/modules/generated/sklearn.compose.make_column_selector.html): Pay attention to the `dtype_include` parameter.

````python
from sklearn.compose import make_column_selector
import numpy as np

numeric_selector = None
# TODO: Instantiate numeric_selector to include only numeric data types
# YOUR CODE HERE
numeric_selector = make_column_selector(dtype_include=np.number)

# Test numeric_selector - it is a function that takes in a DataFrame
# and returns only desired column names
numeric_selector(X_train)
````

Now we need a way to drop the non-numeric columns. There are many ways to do this, but we want to use a ColumnTransformer

Relevant documentation:
* [`make_column_transformer`](https://scikit-learn.org/stable/modules/generated/sklearn.compose.make_column_transformer.html)

````python
from sklearn.compose import make_column_transformer

# We create a transformer that takes in our dataset and keeps only the numeric columns (identified by our numeric selector)
numeric_only_transformer = make_column_transformer(
    ('passthrough', numeric_selector)
)
# Notice the shape of the output has changed, containing only 26 columns
numeric_only_transformer.fit_transform(X_train).shape
````

**Instructions**: Finally, we can combine our `numeric_only_transformer` with our `knn` model into a single pipeline.

Pipelines are useful because they can be trained (`fit`) on the training data or used for inference (`predict`) on the test data. 

For our pipeline, in both cases, the data will go through two steps:
1. The non-numeric columns are be dropped
2. The data is passed into the knn model for either training or prediction

While we could have manually dropped non-numeric columns from our dataset, the pipeline approach is powerful because:
* We can easily change what preprocessing steps we are doing, without modifying our original dataset.
* We can apply the pipeline directly to data that has not been preprocessed (e.g. newly collected data).

We will learn more about pipelines in the next milestone.

Relevant documentation:
* [`make_pipeline`](https://scikit-learn.org/stable/modules/generated/sklearn.pipeline.make_pipeline.html)

This helpfer function returns a DataFrame with the training and evaluation (e.g. test) accuracy for a list of `pipelines`, passed in as the first argument, using a given set of: `X_train`, `y_train`, `X_evaluate`, `y_evaluate`.

**Note**: The input is a list of (name, pipeline) tuples, e.g.
```
evaluate_pipelines(
    [('name1', pipeline_1), ('name2', pipeline_2)],
    X_train, y_train, X_test, y_test
)
```

````python
def evaluate_pipelines(pipelines, X_train, y_train, X_evaluate, y_evaluate):
    rows = []
    for name, pipeline in pipelines:
        pipeline.fit(X_train, y_train)
        training_accuracy = pipeline.score(X_train, y_train)
        test_accuracy = pipeline.score(X_evaluate, y_evaluate)
        rows.append({'name': name, 'training_accuracy': training_accuracy, 'eval_accuracy': test_accuracy})
    return pd.DataFrame(rows)
````

````python
# This makes it easier to get the values we calculated earlier
evaluate_pipelines([('knn', knn_pipeline)], X_train, y_train, X_test, y_test)
````
